I. MODULATION ET DEMODULATION DSB-SC

A. MODULATION
- Fabriquer le signal à transmettre $x(t)$, qui doit être à bande limitée $[-B,B]$ (utiliser éventuellement une fentre adaptée, par exemple de type Kaiser, pour éviter le "spectral leakage"); afficher le signal et son spectre
- Fabriquer le signal modulé $x_m(t)= x(t)cos(2 \pi f_c t)$; afficher le signal et son spectre. Pourquoi le nom "DSB-SC" donné à ce type de modulation?

```python
import numpy as np
#import soundfile as sf
import matplotlib.pyplot as plt
import scipy.signal as sig

B = 0.5
f_c = 50
N = 1000
dt = 0.1
```

```python
T = np.linspace(-6, 6, N)
freqs = np.arange(1000)/1000*B

x = np.cos(2*np.pi*B*T) + 2*np.cos(2*np.pi*2.4*B*T + 3)
x_m = np.cos(2*np.pi*f_c*T)*x

#affichage du signal initial
plt.plot(T,x)
plt.plot(T,x_m)
plt.title('signaux initiaux')
plt.show()

#affichage du signal modulé 
plt.plot(freqs[:N//2], np.abs(np.fft.fft(x)[:N//2]))
plt.plot(freqs[:N//2], np.abs(np.fft.fft(x_m)[:N//2]))
plt.title('spectres')
plt.show()
```

On cherche à comprendre l'appelation "DSB-SC" (Double-SideBand Supressed-Carrier) donnée à ce type de modulation. On obeserve sur le spectre ci dessus que lorsque l'on module le signal (spectre orange), on dédouble son signal: on obtient un "upper side band" et un "lower side band", ce qui explique le "Double sideBand". Ces deux bandes contiennent par ailleurs la même information, ce qui explique qu'on puisse n'en garder qu'une, ce qui se traduit par "supressed-carrier". 

B. DEMODULATION SYNCHRONE "ANALOGIQUE"

- Fabriquer le signal $x_{md}(t)=x_m(t)cos(2 \pi f_c t)$; afficher son signal et son spectre
- Filtrer $x_{md}(t)$ par un filtre basse-bas, soit $x_{mdf}(t)$ le résultat; afficher le signal et son spectre. Conclusion?
- Pour la suite du traitement, on échantillonnerait $x_{mdf}(t)$ afin de manipuler numériquement le signal. Quelle est la fréquence minimale d'échantillonage possible?
- La démodulation présentée ici se fait d'habitude avec des composants analogiques. Serait-il possible de réaliser numériquement le même traitement après avoir échantillonné $x_m$? Quelle est la fréquence d'échantillonage minimale possible ?

```python
x_md = np.cos(2*np.pi*f_c*T)*x_m

#affichage du signal et du spectre
plt.plot(T,x_md)
plt.title('signal')
plt.show()
plt.plot(freqs[:N//2], np.abs(np.fft.fft(x_md)[:N//2]))
plt.title('spectre')
plt.show()
```

```python
def FiltrePB(y, dt, N_filter, fc):
    num_low, den_low = sig.butter(N_filter, fc, btype='lowpass', fs=1/dt)
    y_low = sig.filtfilt(num_low, den_low, y)
    return y_low

x_mdf = FiltrePB(x_md, 0.1, 2, 2*B)

#affichage du signal et du spectre
plt.plot(T, x_mdf)
plt.title('signal')
plt.show()
plt.plot(freqs[:N//2], np.abs(np.fft.fft(x_mdf)[:N//2]))
plt.title('spectre')
plt.show()
```

On se rend compte, en comparant tout d'abord les deux signaux précédents, que les composantes rapides ont été supprimées, ce qui est cohérent avec l'utilisation d'un passe-bas. On peut faire la même remarque lorsque l'on compare les spectres. 

Par ailleurs, en regardant le spectre initial -non modulé-, on se rend compte qu'une telle modulation puis l'application d'un filtre passe-bas permet de récupérer le signal d'origine.
